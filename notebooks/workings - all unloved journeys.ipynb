```python
clean_content_exploded_df = pd.read_csv('/Users/suganyasivaskantharajah/code/govuk_ab_analysis/clean_content_with_rl_exploded.csv.gz')
```

```python
# dedupe the pagePaths here just in case
loved_page_paths = list(set(clean_content_exploded_df['pagePath'].tolist()))
```

```python
len(loved_page_paths)
```

```python
one_day_df['Page_List'] = one_day_df['Page_List'].progress_apply(ast.literal_eval)
```

```python
def is_loved_journey(page_list):
    unique_pages = list(set(page_list))
    return any([page in loved_page_paths for page in unique_pages])

def is_loved_journey_2(page_list):
#     we dedupe each pagelist here
    return any([page in loved_page_paths for page in set(page_list)])

def is_loved_journey_3(page_list):
    return any([page in loved_page_paths for page in list(set(page_list))])

def is_loved_journey_4(page_list):
    return any([page in set(loved_page_paths) for page in list(set(page_list))])
```

```python
sample_df = one_day_df.sample(10000)
```

```python
%timeit sample_df['Page_List'].apply(is_loved_journey)
```

```python
%timeit sample_df['Page_List'].apply(is_loved_journey_2)
```

```python
%timeit sample_df['Page_List'].apply(is_loved_journey_3)
```

```python
%timeit sample_df['Page_List'].apply(is_loved_journey_4)
```

```python
loved_page_paths_set = set(loved_page_paths)

def is_loved_journey_5(page_list):
    return any([page in loved_page_paths_set for page in list(set(page_list))])
```

```python
%timeit sample_df['Page_List'].apply(is_loved_journey_5)
```

```python
one_day_df['is_loved_journey'] = one_day_df['Page_List'].progress_apply(is_loved_journey)
```
